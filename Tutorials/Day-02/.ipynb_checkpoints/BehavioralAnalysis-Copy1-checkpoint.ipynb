```python
import warnings
warnings.simplefilter(action='ignore', category=Warning)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy
from pathlib import Path
from glob import glob
import math
import statsmodels.api as sm
# from statsmodels.formula.api import ols
import statsmodels.formula.api as smf
from statsmodels.stats.anova import AnovaRM
import scipy.stats as spstat
DATA_PATH = './Behavioral_data_robot/'# check path before running
```

## Read ball position data

```python
positions_df = pd.DataFrame() 
for fname in Path(DATA_PATH).rglob('*.csv'):
    tmpdf = pd.read_csv(fname, sep=',', header=0, skipinitialspace=True, index_col=False)
    
    tmpdf['fname'] = str(fname)
    
    tmpdf.columns = tmpdf.columns.str.lstrip().str.replace(' ', '_')
    
    positions_df = pd.concat([positions_df, tmpdf], ignore_index=True, sort=False, axis=0)

positions_df['session'] = positions_df.fname.str.extract(r".*ses([0-9]).*")
positions_df['subject'] = positions_df.fname.str.extract(r".*sub([0-9])_robot.*")
positions_df['sono'] = positions_df.fname.str.extract(r".*sono([0-9]).*")
positions_df['partner'] = positions_df.fname.str.extract(r'.*(robot).*')


positions_df.partner.fillna('human', inplace=True)
positions_df.subject.fillna('3', inplace=True)
positions_df.dropna(subset=['sono'], inplace=True)
# positions_df.columns =  positions_df.columns.str.lstrip().str.replace(' ', '_')
```

```python
positions_df.info() # get info dataframe columns
```

```python
final_df = (
    positions_df
    .query('session != "5" and session != "7" and TIME>=0')
) #remove noisy session and wrong timestamps

final_df['theta'] = np.arctan2(final_df.BALL_Y, final_df.BALL_X)

# convert range from [-pi, pi] to [0, 2pi]
final_df['theta_radians'] = (final_df['theta'] + 2 * np.pi) % (2 * np.pi) 
final_df['theta_degrees'] = (180/np.pi) * final_df['theta_radians']

final_df['sono_'] = final_df.sono.map({'0':'OFF', '1':'ON'})

display(final_df.head())
```

## calculate velocity and acceleration of the ball

```python
final_df['vel_theta'] = (final_df
                        .groupby(['session', 'subject', 'partner', 'fname']).theta_degrees
                        .apply(lambda x: x.diff())
                       )/(final_df
                        .groupby(['session', 'subject', 'partner', 'fname']).TIME
                        .apply(lambda x: x.diff())
                       )
final_df.vel_theta = final_df.vel_theta.abs()

final_df['acc_theta'] = (final_df
                        .groupby(['session', 'subject', 'partner', 'fname']).vel_theta
                        .apply(lambda x: x.diff())
                       )/(final_df
                        .groupby(['session', 'subject', 'partner', 'fname']).TIME
                        .apply(lambda x: x.diff())
                       )
final_df.dropna(subset=['vel_theta'], inplace=True)
```

## calculate deviation of ball from circular track

```python
final_df['error'] = np.sqrt(final_df['BALL_X']**2 + final_df['BALL_Y']**2) -  0.67

display(final_df.head())
```

## describe velocity per partner and sonification

```python
mean_velocity_per_session = (
    final_df
    .groupby(['session','subject','partner', 'sono_'])
    .vel_theta
    .mean()
    .reset_index(name='mean_velocity')
)

mean_velocity_per_session.to_csv('mean_velocity_session.csv', index=False)
```

```python
## dulicate velocity for each dyad in session in partner:human condition
tmpdf = mean_velocity_per_session
partner_human_rows = tmpdf.query('partner=="human"')
tmpdf.loc[tmpdf.partner=='human', 'subject'] = '1'
partner_human_rows.loc[partner_human_rows.partner=='human', 'subject'] = '2'
partner_df = pd.DataFrame()
partner_df = pd.concat([partner_human_rows, tmpdf], ignore_index=True)
partner_df['id'] = partner_df.session.astype(int) * (partner_df.subject.astype(int) + 1000)

display(partner_df.groupby(['partner']).describe())
display(partner_df.groupby(['sono_']).describe())
# df.sort_values(by='session')
```

```python
_, ax = plt.subplots(figsize=(7,7), sharey=True, sharex=True)
# ax = ax.ravel()
sns.set(context='poster', style='ticks', font_scale=1., palette='muted')

ax = sns.pointplot(
    data=partner_df,
    x='partner',
    y='mean_velocity',
    hue='sono_',
    order=['robot', 'human'],
    ci=68,
    palette=['b', 'r'],
    dodge=0.2,   
    capsize=0.1,
    errwidth=2.5,
    join=False
)
ax.set(
    ylabel='Mean Velocity [radians/sec]',
    xlabel='Partner',
    ylim=(140, 300),
    yticks=np.arange(50,301,50)
)
ax.legend(frameon=False, title='Sonification', bbox_to_anchor=(1.1,0.26))
sns.despine(offset=10)
# plt.savefig('mean_velocity.eps', transparent=True, bbox_inches='tight')
```

## ANOVA for differences in angular velocity 
within subject factors partner and sonification

```python
anova = AnovaRM(partner_df, 'mean_velocity', 'id', within=['partner', 'sono_']).fit()
print(anova.anova_table)
```

## describe error rates per condition

```python
mean_error_per_session = (
    final_df
    .groupby(['session','subject','partner', 'sono_'])
    .error
    .mean()
    .reset_index(name='mean_error')
)
mean_error_per_session.to_csv('mean_error_session.csv', index=False)
```

```python
## dulicate error for each dyad in session in partner:human condition
tmpdf = mean_error_per_session
partner_human_rows = tmpdf.query('partner=="human"')
tmpdf.loc[tmpdf.partner=='human', 'subject'] = '1'
partner_human_rows.loc[partner_human_rows.partner=='human', 'subject'] = '2'
sono_df = pd.DataFrame()
sono_df = pd.concat([partner_human_rows, tmpdf], ignore_index=True)
sono_df['id'] = sono_df.session.astype(int) * (sono_df.subject.astype(int) + 1000)
# df.sort_values(by='session')
display(sono_df.groupby(['partner']).describe())
display(sono_df.groupby(['sono_']).describe())
```

```python
_, ax = plt.subplots(figsize=(7,7), sharey=True, sharex=True)
# ax = ax.ravel()
sns.set(context='poster', style='ticks', font_scale=1., palette='muted')

ax = sns.pointplot(
    data=sono_df,
    x='partner',
    y='mean_error',
    hue='sono_',
    order=['robot', 'human'],
    ci=68,
    palette=['b', 'r'],
    dodge=0.2,
    capsize=0.1,
    errwidth=2.5,
    join=False
)
ax.set(
    ylabel='Mean Error',
    xlabel='Partner',
    ylim=(0.,0.06),
#     yticks=[2,3,4,5]
)
ax.legend(frameon=False, title='Sonification', bbox_to_anchor=(1.1,0.26))
sns.despine(offset=10)
plt.savefig('mean_error.eps', transparent=True, bbox_inches='tight')
```

## ANOVA for differences in errors
within subject factors partner and sonification

```python
anova = AnovaRM(sono_df, 'mean_error', 'id', within=['partner', 'sono_']).fit()
print(anova.anova_table)
```

## Correlation between mean angular velocity and error

```python
df = pd.concat(
    [partner_df.set_index(['id', 'session', 'subject', 'partner', 'sono_', ]), 
    sono_df.set_index(['id', 'session', 'subject', 'partner', 'sono_', ])]
    , axis=1
    
)
display(df.reset_index().sort_values(by='id'))
```

```python
# HUMAN v OFF
print(
    scipy.stats.pearsonr(
        df.query('partner=="human" and sono_=="OFF"').mean_velocity, 
        df.query('partner=="human" and sono_=="OFF"').mean_error)
)
# HUMAN v ON
print(
    scipy.stats.pearsonr(
        df.query('partner=="human" and sono_=="ON"').mean_velocity, 
        df.query('partner=="human" and sono_=="ON"').mean_error)
)
# ROBOT v OFF
print(
    scipy.stats.pearsonr(
        df.query('partner=="robot" and sono_=="OFF"').mean_velocity, 
        df.query('partner=="robot" and sono_=="OFF"').mean_error)
)
# ROBOT v ON
print(
    scipy.stats.pearsonr(
        df.query('partner=="robot" and sono_=="ON"').mean_velocity, 
        df.query('partner=="robot" and sono_=="ON"').mean_error)
)
```

```python
ax = sns.lmplot(
    data=tmpdf, x='mean_velocity', y='mean_error', 
    hue='partner', col='sono_', 
    height=7, aspect=1, truncate=0, palette='muted',
    x_jitter=2, y_jitter=0.01
)
ax.set(
    ylim=(0,0.08),
    xlim=(100, 300),
    yticks=np.arange(0,0.09, 0.02)
)

plt.savefig('corr_velocity_error.eps', bbox_inches='tight', quality=90)
```

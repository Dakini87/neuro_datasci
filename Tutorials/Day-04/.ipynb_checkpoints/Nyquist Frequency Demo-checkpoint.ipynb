```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from wand.image import Image
from wand.display import display
```

```python
def createSignal(freqList, lenSignal, timePeriod, burstFreq, burstTime):
    t = linspace(timePeriod[0], timePeriod[1], lenSignal)
    signal =  np.sin(np.pi * 2. * burstFreq * t)    
    signal[np.logical_or(t < burstTime[0], t > burstTime[1])] = 0
    on_time = np.logical_and(t >= burstTime[0], t <= burstTime[1])
    signal[on_time] *= np.hanning(on_time.sum()) 
    for f in freqList:
        signal += np.sin(2*np.pi*f*t)
        
    return signal


def getFT(signal):
    ft = np.fft.fft(signal)/len(signal)
    return np.abs(ft[:len(signal)//2])
```

```python
# %matplotlib notebook
def getFT(signal):
    ft = fft.fft(signal/len(signal))
    return abs(ft[:len(signal)//2])

def clt(current):
    # if animation is at the last frame, stop it
    current = current+10
    plt.cla()
    if current == 160: 
        a.event_source.stop()
    # sampling a signal badlimited to 40 Hz 
    # with a sampling rate of 800 Hz
    f = 40;  # Hz
    tmin = -0.5;
    tmax = 0.5;
    t = linspace(tmin, tmax, 400);
    x = cos(2*pi*t) + cos(2*pi*f*t); # signal 
    ft_x = getFT(x)
    plt.plot(ft_x, 'g', label='Original')   
    # sampling the signal with a sampling rate of 0.5 of signal frequency
    # in this case, we are using the Nyquist rate.
    T = 1/80;
    nmin = ceil(tmin / T);
    nmax = floor(tmax / T);
    n = arange(nmin,nmax);
    x1 =  cos(2*pi*n*T) + cos(2*pi*f*n*T);
#     plt.plot(n*T, x1, '-b', label='Inline label')
    # sampling the signal with a sampling rate of 35 Hz
    # note that 35 Hz is under the Nyquist rate.
    T = 1/current;
    nmin = ceil(tmin / T);
    nmax = floor(tmax / T);
    n = arange(nmin,nmax);
    x2 = cos(2*pi*n*T) + cos(2*pi*f*n*T);
    ft_x2 = getFT(x2)
    plt.plot(ft_x2, '--r', label='Reconstructed')
    
    plt.axis([0, 100, 0, 1])
    plt.gca().set_title('Sampling the signal with frequency = {} Hz '.format(current))
    plt.gca().set_xlabel('Frequency')
    plt.legend()
    
fig = plt.figure(figsize=(8, 6))
a = animation.FuncAnimation(fig, clt, interval=1000)
```

```python
%matplotlib inline
freqList = [ 35, 50]
N = 1024
timePeriod = [0,1]
burstFreq = 15
burstTime = [0.2, 0.6]
x = createSignal(freqList, N, timePeriod, burstFreq,burstTime)
x_ft = (getFT(x))**2

fig, axes = plt.subplots(2, 1, figsize=(8,6))
axes[0].plot(x)
axes[0].set(xlabel='Time')
axes[1].plot(x_ft)
axes[1].set(xlabel='Frequency', xlim=(0,150), ylabel='Power (\muV^2)')
```

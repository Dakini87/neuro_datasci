```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
group_names = ['angelfish-of-sublime-enrichment',
               'adamant-strong-gecko',
               'ambitious-futuristic-ibis',
               'beautiful-mosquito-of-perception',
#                'fair-flying-newt',
               'first-granite-leopard',
#                'industrious-nonchalant-camel',
               'invaluable-wildebeest-of-blizzard',
               'knowing-garnet-bear',
               'large-tall-ostrich',
#                'masterful-vole-of-defiance',
               'messy-marvellous-antelope',
               'naughty-quartz-cat',
#                'placid-dazzling-squid',
               'valiant-dynamic-wolverine',
               'vivid-adorable-salamander'
]
```

```python
group_df = pd.DataFrame()
group_df['names'] = group_names
group_df['bonus_count'] = 1
group_df['day'] = 1
```

```python
fig, ax = plt.subplots(figsize=(10,20))
sns.set(style='ticks', context='talk', font_scale=2, )
sns.barplot(
    data=group_df, 
    y='names', x='bonus_count', 
    order=group_df.names.sort_values(), 
    palette='icefire',
    ax=ax
);

ax.set(
    xlim=(0,5),
    ylabel='',
    xlabel='# of bonus exercises'
);
```
